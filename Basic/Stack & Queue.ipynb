# STACK 

- LIFO (Last-in, First-out)
- Stack 은 연속으로 저장된 데이터 구조를 갖고 있고, ***맨 위 요소에 대한 포인터(주소값)*** 을 저장하고 있는 Array or Sigly Linked List 로 구현

****
- 장점 : 참조 지역성 (한번 참조된 곳은 다시 참조될 확률이 높음)
- 단점 : 데이터 탐색이 어려움

****
- 사용 예시
    - 문자열 역순 출력, 연산자 후위표기법
****
- ADT (abstract data type)

    - ***push*** (return None): (필수구현) 맨 위에 값 추가
    - ***pop***  (return data): (필수구현) 맨 위의 값 제거
    - peak       (return data or -1): 스택 변형 없이 맨 위의 값 출력
    - is_empty   (return boolean): 스택이 비어있는지 확인
****

## Stack with List

```python
class Stack(list):
    def __init__(self):
        self.stack = [] # 리스트 초기화
    
    def push(self, data):
        self.stack.append(data) # 데이터 추가 -> append 사용시 마지막에 추가되니까
    
    def pop(self):
        if self.is_empty():
            return -1
        return self.stack.pop()
    
    def peek(self):
        return self.stack[-1]
    
    def is_empty(self):
        if len(self.stack) == 0:
            return True
        return False
    
if __name__ == "__main__":
    s = Stack()
    s.push(1)
    s.push(2)
    s.push(3)
    stack = s.stack
    
    print("print stack")
    for i in stack:
        print(i)
        
    print("peek", s.peek())
    print("pop", s.pop())
    print("pop", s.pop())
```

## Stack with LinkedList

- ***Top*** :마지막 데이터를 가리키는 변수

```python
Image("../../img/stack_linkedlist.png") # code안에서 나오게 할 때
```

```python
# linkedlist
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Stack
class Stack:
    def __init__(self):
        self.head = None
    
    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def pop(self):
        if self.is_empty():
            return -1
        data = self.head.data
        self.top = self.head.next
        return data
    
    def is_empty(self):
        if self.head:
            return False
        return True
    
    def peek(self):
        if self.is_empty():
            return -1
        return self.head.data
    
```

# Queue

- FIFO (First-in, First-out)
- 데이터가 ***입력한 순서대로 처리되어야 할 경우*** 사용
- 리스트로 구현할 경우 끝에 데이터를 붙이거나, 끝에서 데이터를 꺼내는 작업은 빠르지만 맨 앞의 데이터를 추가, 빼는 것은 느림 (다른 데이터들 을 한칸씩 이동시켜야 하기 때문)
- library (collections.deque, queue) or linkedlist 사용
****
- 종류
    - 선형 Queue
    - 원형 Queue
    - 우선순위 Queue

****
- 사용 예시
    - 문자열 역순 출력, 연산자 후위표기법
****
- ADT (abstract data type)
    - front(head) : 데이터를 dequeue 할 수 있는 위치
    - rear(tail) : 데이터를 enqueue 할 수 있는 위치
   
    - ***enqueue*** : insert
    - ***dequeue*** :  delete
    - peek : 맨 앞의 값 반환만 하고 삭제는 안함
    - is_empty()

## Queue with List
- enqueue : O(1)
- dequeue : O(N) ==> 맨 앞의 데이터 삭제시, 두번째 데이터부터 마지막 데이터까지 왼쪽으로 한칸씩 옮겨야 함

```python
class Queue(object):
    def __init__(self):
        self.queue = []
    
    def dequeue(self):
        if len(self.queue) == 0:
            return -1
        return self.queue.pop(0)
    
    def enqueue(self, n):
        self.queue.append(n)
        
    def printQueue(self):
        print(self.queue)
        

if __name__ == "__main__":
    q = Queue()
    q.enqueue(1)
    q.enqueue(2)
    q.enqueue(3)
    q.enqueue(4)
    
    print("Before dequeue")
    q.printQueue()
    
    print(q.dequeue())
    
    print("After dequeue")
    q.printQueue()
        
```

## Queue with LinkedList

```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Queue(object):
    def __init__(self):
        self.head = None
        self.tail = None
    
    def enqueue(self, node):
        if self.head == None:
            self.head = node
            self.tail = node
        else:
            self.tail.next = node
            self.tail = self.tail.next
            
    def dequeue(self):
        if self.head == None:
            return -1
        
        v = self.head.data
        self.head = self.head.next
        
        if self.head == None:
            self.tail = None
            
        return v
    
    def printQueue(self):
        check = self.head
        string = ""
        while check:
            string += str(check.data)
            if check.next:
                string += "->"
            check = check.next
        print(string)
        
if __name__ == "__main__":
    q = Queue()
    q.enqueue(Node(1))
    q.enqueue(Node(2))
    q.enqueue(Node(3))
    
    q.printQueue()
    
    print(q.dequeue())
    print(q.dequeue())
    
    q.printQueue()
```

### deque (double-ended queue)

deque : 앞과 뒤 양방향에서 데이터를 처리할 수 있는 자료구조
-  앞,뒤에서 데이터를 처리하는 속도가 O(1)
- (내부적으로 double-linkedlist 로 구현되어 있음)

```python
from collections import deque

dq = deque([])

dq.append(1)
dq.append(2)
dq.append(3)

print("1. dq : ", dq)

print("2. popLeft : ", dq.popleft())
print("3. after popLeft :", dq)

print("pop : ", dq.pop())
print("after pop :", dq)


print("4. append", dq.append(10)) # pop 
print("5. after append : ", dq)

print("4. append", dq.appendleft(20)) # pop 
print("5. after appendleft : ", dq)


```
