Given an array nums of n integers, are there elements a, b, c in nums such that a + b + c = 0? Find all unique triplets in the array which gives the sum of zero.

Notice that the solution set must not contain duplicate triplets.

*Constraints:*
- 0 <= nums.length <= 3000
- -10^5 <= nums[i] <= 105

```python
from itertools import combinations

class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        answer = []
        combis = list(combinations(nums, 3))
        
        for combi in combis:
            combi = sorted(combi)
            if combi not in answer and sum(combi) == 0:
                answer.append(list(combi))
        return answer
                
```

# Solution1 : ë¸Œë£¨íŠ¸ í¬ìŠ¤ë¡œ ê³„ì‚° 
- ì•, ë’¤ë¡œ ê°™ì€ ê°’ì´ ìˆì„ ê²½ìš° ì´ë¥¼ ì‰½ê²Œ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ ë¨¼ì € sortí•¨ìˆ˜ë¡œ ì •ë ¬
    - pythonì˜ sort í•¨ìˆ˜ :íŒ€ì†ŒíŠ¸ --> **O(nlogn)**
- i, j, k ì„¸ê°œì˜ í¬ì¸í„°ê°€ ê³„ì† ì´ë™í•˜ë©´ì„œ i+j+k=0 ì°¾ì•„ë‚´ëŠ” ë¸Œë£¨íŠ¸ í¬ìŠ¤ ë°©ì‹

```python
class Solution:
    def threeSum(self, nums):
        results = []
        nums.sort()

        # ë¸Œë£¨íŠ¸ í¬ìŠ¤ n^3 ë°˜ë³µ
        for i in range(len(nums) - 2):
            # ì¤‘ë³µëœ ê°’ ê±´ë„ˆë›°ê¸°
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            for j in range(i + 1, len(nums) - 1):
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue
                for k in range(j + 1, len(nums)):
                    if k > j + 1 and nums[k] == nums[k - 1]:
                        continue
                    if nums[i] + nums[j] + nums[k] == 0:
                        results.append([nums[i], nums[j], nums[k]])

        return results
```

```python
class Solution:
    def threeSum(self, nums):
        results = []
        nums.sort()

        for i in range(len(nums) - 2):
            # ì¤‘ë³µëœ ê°’ ê±´ë„ˆë›°ê¸°
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            # ê°„ê²©ì„ ì¢í˜€ê°€ë©° í•© `sum` ê³„ì‚°
            left, right = i + 1, len(nums) - 1
            while left < right:
                sum = nums[i] + nums[left] + nums[right]
                if sum < 0:
                    left += 1
                elif sum > 0:
                    right -= 1
                else:
                    # `sum = 0`ì¸ ê²½ìš°ì´ë¯€ë¡œ ì •ë‹µ ë° ìŠ¤í‚µ ì²˜ë¦¬
                    results.append([nums[i], nums[left], nums[right]])

                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
                    left += 1
                    right -= 1

        return results
```
